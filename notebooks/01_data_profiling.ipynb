## 2 - Missing Data Strategy Design
### 2.1 Missing Value Audit

```python
df.isnull().sum()
```

```python
df["isFraud"].mean()
```

```python
# Check if target exists
if "isFraud" in df.columns:
    print("Target column found ✅")
else:
    print("Target column NOT found ❌")

# Class distribution
fraud_counts = df["isFraud"].value_counts()

print("\nFraud Distribution:")
print(fraud_counts)

imbalance_ratio = fraud_counts[0] / fraud_counts[1]
print("Imbalance Ratio:", imbalance_ratio, "\n")

print(df["isFraud"].value_counts(normalize=True))  # normalize=True gives percentage distribution

# Percentage distribution
fraud_percent = df["isFraud"].value_counts(normalize=True) * 100  

print("\nFraud Percentage:")
print(fraud_percent)
```

### 2.2 Build Missing Summary Table

```python
missing_summary = pd.DataFrame({
    "missing_count": df.isnull().sum(),
    "missing_percent": df.isnull().mean() * 100,
    "dtype": df.dtypes
}).sort_values("missing_percent", ascending=False)

missing_summary
```

### 2.3 Test Missing Signal Strength

```python
high_missing_cols = missing_summary[
    missing_summary["missing_percent"] > 99
].index.tolist()

signal_results = {}

for col in high_missing_cols:
    fraud_rate = df.groupby(df[col].isnull())["isFraud"].mean()

    if len(fraud_rate) == 2:  # ensure both True/False exist
        signal_strength = abs(fraud_rate[True] - fraud_rate[False])
        signal_results[col] = signal_strength

# Convert to DataFrame
signal_df = pd.DataFrame.from_dict(
    signal_results,
    orient="index",
    columns=["fraud_rate_difference"]
).sort_values("fraud_rate_difference", ascending=False)

signal_df.head()
```

## fraud_rate[True]

Fraud rate when the column value is **missing (NULL)**:

$$
fraud\_rate[True] =
\frac{Fraud\ Cases\ (col\ is\ NULL)}
{Total\ Rows\ (col\ is\ NULL)}
$$

---

## fraud_rate[False]

Fraud rate when the column value is **NOT missing (NOT NULL)**:

$$
fraud\_rate[False] =
\frac{Fraud\ Cases\ (col\ is\ NOT\ NULL)}
{Total\ Rows\ (col\ is\ NOT\ NULL)}
$$


## Signal Strength Equation

Signal strength measures the absolute difference between fraud rates
when the column is missing and when it is not missing.

$$
signal\_strength = 
\left| 
fraud\_rate[True] - fraud\_rate[False] 
\right|
$$

             A small difference → weak signal
             A large difference → strong signal
